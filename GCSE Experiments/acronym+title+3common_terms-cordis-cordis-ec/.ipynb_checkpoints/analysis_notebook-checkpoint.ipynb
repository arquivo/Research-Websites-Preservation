# Using Fuzzy Hashing to compare contents

```python
# SKIP THIS STEP, ALREADY AS BEEN RUN.
# JUST LOAD THE DATA FROM THE TXT FILE fuzzy_computation_backup.txt

import urllib2
import ssdeep

df['projectUrlHash'] = np.nan
df['foundProjectUrl1Hash'] = np.nan
df['foundProjectUrl2Hash'] = np.nan
df['foundProjectUrl3Hash'] = np.nan
df['foundProjectUrl4Hash'] = np.nan
df['foundProjectUrl5Hash'] = np.nan
df['foundProjectUrl6Hash'] = np.nan
df['foundProjectUrl7Hash'] = np.nan
df['foundProjectUrl8Hash'] = np.nan
df['foundProjectUrl9Hash'] = np.nan
df['foundProjectUrl10Hash'] = np.nan
df['MatchScore1'] = np.nan
df['MatchScore2'] = np.nan
df['MatchScore3'] = np.nan
df['MatchScore4'] = np.nan
df['MatchScore5'] = np.nan
df['MatchScore6'] = np.nan
df['MatchScore7'] = np.nan
df['MatchScore8'] = np.nan
df['MatchScore9'] = np.nan
df['MatchScore10'] = np.nan



for index,row in df.iterrows():    
    print "computing fuzzy hash for project %s" % row['acronym']
    try:
        df.ix[index, 'projectUrlHash'] = ssdeep.hash(urllib2.urlopen(row['projectUrl']).read())
    except urllib2.HTTPError, e:
        print  e.code               
    except urllib2.URLError, e:        
        print e.reason
    except Exception:
        print 'Unhandled Error'     
    for i in np.arange(1,11):
        try:            
            df.ix[index, 'foundProjectUrl'+ str(i) +'Hash'] = ssdeep.hash(urllib2.urlopen(row['foundProjectUrl' + str(i)]).read())         
            df.ix[index, 'MatchScore' + str(i)] = ssdeep.compare(df.ix[index,'projectUrlHash'],df.ix[index,'foundProjectUrl' + str(i)+ 'Hash'])
        except urllib2.HTTPError, e:
            print  "Error at foundProjectUrl" + str(i) + " %s" % e.code               
        except urllib2.URLError, e:        
            print "Error at foundProjectUrl" + str(i) + " %s" % e.reason
        except Exception:
            print 'Unhandled Error at foundProjectUrl' + str(i)       
```

```python
df = pd.read_csv('fuzzy_computation_backup.csv', sep =',', quotechar = '"', index_col=False)
```

### Matching result with Fuzzy hashing comparison

```python
total_entries = df['acronym'].count()
exact_match_count_top1_fuzzy = df['acronym'][(df['projectUrl'] == df['foundProjectUrl1']) | (df['MatchScore1'] >= 80)].count()
print "Number of exact URL matches or Fuzzy Hashing 100%% Match: %s of %s" % (exact_match_count_top1, total_entries)
print "Matching: %s%%" % (exact_match_count_top1_fuzzy * 100 / total_entries)

exact_match_count_top3_fuzzy = total_entries = df['acronym'].count()
exact_match_count_top3 = df['acronym'][(df['projectUrl'] == df['foundProjectUrl1'] ) | 
                                       (df['projectUrl'] == df['foundProjectUrl2']) | 
                                       (df['projectUrl'] == df['foundProjectUrl3']) |
                                       (df['MatchScore1'] >= 80) |
                                       (df['MatchScore2'] >= 80) |
                                       (df['MatchScore3'] >= 80)].count()
print "Number of exact matches: %s of %s" % (exact_match_count_top3,total_entries)
print "Matching: %s%%" % (exact_match_count_top3 * 100 / total_entries)
```

```python
matching_top1 = []
matching_top3 = []
matching_top5 = []
matching_top10 = []
for i in np.arange(40,110,10):
    matching_top1.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1']) | (df['MatchScore1'] >= i)].count() / float(total_entries))
    matching_top3.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1']) | (df['MatchScore1'] >= i) |
                                       (df['MatchScore2'] >= i) |
                                       (df['MatchScore3'] >= i)].count() / float(total_entries))
    matching_top5.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1']) | (df['MatchScore1'] >= i) |
                                       (df['MatchScore2'] >= i) |
                                       (df['MatchScore3'] >= i) |
                                       (df['MatchScore4'] >= i) |
                                       (df['MatchScore5'] >= i)].count() / float(total_entries))  
    matching_top10.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1']) | (df['MatchScore1'] >= i) |
                                       (df['MatchScore2'] >= i) |
                                       (df['MatchScore3'] >= i) |
                                       (df['MatchScore4'] >= i) |
                                       (df['MatchScore5'] >= i) |
                                       (df['MatchScore6'] >= i) |
                                       (df['MatchScore7'] >= i) |
                                       (df['MatchScore8'] >= i) |
                                       (df['MatchScore9'] >= i) |
                                       (df['MatchScore10'] >= i)].count() / float(total_entries))      
plt.plot(np.arange(40,110,10),matching_top1,linestyle='--',color='r',marker='o',label='Top 1 Results')
plt.plot(np.arange(40,110,10),matching_top3,linestyle='--',color='b',marker='o',label='Top 3 Results')
plt.plot(np.arange(40,110,10),matching_top5,linestyle='--',color='y',marker='o',label='Top 5 Results')
plt.plot(np.arange(40,110,10),matching_top10,linestyle='--',color='g',marker='o',label='Top 10 Results')
plt.plot((40,100),(0.43,0.43),'-r')
plt.plot((40,100),(0.51,0.51),'-b')
plt.plot((40,100),(0.55,0.55),'-y')
plt.plot((40,100),(0.56,0.56),'-g')

plt.xlabel("Similarity Score")
plt.ylabel("Result Match Success")
plt.grid()
plt.legend()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
