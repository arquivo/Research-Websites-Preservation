```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys, os, re
%matplotlib inline
```

```python
df = pd.read_csv('cordisfp5noobjectives.csv', sep=';', index_col = False)
df.head()
```

```python
df = df[['Project Acronym','Project Title','Project Website']]
df = df.fillna(0)
```

## Rename csv fields

```python
df.columns = ['acronym','title','projectUrl',]
```

```python
df.head()
```

## Cleaning

```python
df['title'] = df['title'].apply(lambda x : re.sub(r'[\W_]', ' ', x))
```

```python
for index, row in df.iterrows():    
    df.ix[index]['acronym'] = str(row['acronym']).lstrip()
    df.ix[index]['acronym'] = str(row['acronym']).rstrip()
    df.ix[index]['title'] = str(row['title']).lstrip()
    df.ix[index]['title'] = str(row['title']).rstrip()
    
    df.ix[index]['title'] = ' '.join(row['title'].split())    
    if row['projectUrl'] != 0:
        try:
            df.ix[index]['projectUrl'] = row['projectUrl'].lstrip()
            df.ix[index]['projectUrl'] = row['projectUrl'].rstrip()
        except Exception, e:
            print row['projectUrl']
```

```python
# get projects without a URL
df_nourl = df[df['projectUrl'] == 0]
```

```python
df_nourl.to_csv('fp5_nourl_list.txt', index = False)    
```
