# Get results from gcse experiments

```python
# get results from gcse experiments

# acronym + title
header_names = ['acronym','title','projectUrl','foundProjectUrl1',
                'foundProjectUrl2','foundProjectUrl3','foundProjectUrl4',
                'foundProjectUrl5','foundProjectUrl6','foundProjectUrl7',
                'foundProjectUrl8','foundProjectUrl9','foundProjectUrl10',]
df_gcse_acron_title = pd.read_csv('gcse_experiments/acronym+title/output.csv', sep =',', quotechar = '"',names = header_names, index_col=False)
df_gcse_acron_title_cordis = pd.read_csv('gcse_experiments/acronym+title-cordis/output.csv', sep =',', quotechar = '"',names = header_names, index_col=False)
df_gcse_acron_title_cordis_ec = pd.read_csv('gcse_experiments/acronym+title-cordis-ec/output.csv', sep =',', quotechar = '"',names = header_names, index_col=False)

Datasets = []
Datasets.append(df_gcse_acron_title)
Datasets.append(df_gcse_acron_title_cordis)
Datasets.append(df_gcse_acron_title_cordis_ec)

total_entries = df_gcse_acron_title['acronym'].count()
```

### Matching

```python
gcse_matches_top1 = []
gcse_matches_top3 = []
gcse_matches_top5 = []
gcse_matches_top10 = []

for df in Datasets:        
        gcse_matches_top1.append(df['acronym'][df['projectUrl'] == df['foundProjectUrl1']].count())
        
        gcse_matches_top3.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1'] ) | 
                                          (df['projectUrl'] == df['foundProjectUrl2']) |
                                          (df['projectUrl'] == df['foundProjectUrl3'])].count())
        
        gcse_matches_top5.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1'] ) | 
                                          (df['projectUrl'] == df['foundProjectUrl2']) | 
                                          (df['projectUrl'] == df['foundProjectUrl3']) |
                                          (df['projectUrl'] == df['foundProjectUrl4']) |
                                          (df['projectUrl'] == df['foundProjectUrl5'])].count())
        
        gcse_matches_top10.append(df['acronym'][(df['projectUrl'] == df['foundProjectUrl1']) | 
                                          (df['projectUrl'] == df['foundProjectUrl2']) | 
                                          (df['projectUrl'] == df['foundProjectUrl3']) |
                                          (df['projectUrl'] == df['foundProjectUrl4']) |
                                          (df['projectUrl'] == df['foundProjectUrl5']) |
                                          (df['projectUrl'] == df['foundProjectUrl6']) |
                                          (df['projectUrl'] == df['foundProjectUrl7']) |
                                          (df['projectUrl'] == df['foundProjectUrl8']) |
                                          (df['projectUrl'] == df['foundProjectUrl9']) |
                                          (df['projectUrl'] == df['foundProjectUrl10'])].count())
        
        
```

```python
gcse_acron_title_values = [gcse_matches_top1[0]/float(total_entries),gcse_matches_top3[0]/float(total_entries),
                      gcse_matches_top5[0]/float(total_entries),gcse_matches_top10[0]/float(total_entries)]
gcse_acron_title_cordis_values = [gcse_matches_top1[1]/float(total_entries),gcse_matches_top3[1]/float(total_entries),
                      gcse_matches_top5[1]/float(total_entries),gcse_matches_top10[1]/float(total_entries)]
gcse_acron_title_cordis_ec_values = [gcse_matches_top1[2]/float(total_entries),gcse_matches_top3[2]/float(total_entries),
                      gcse_matches_top5[2]/float(total_entries),gcse_matches_top10[2]/float(total_entries)]
```

### Plot

```python
x_values = [1,3,5,10]
plt.plot(x_values,gcse_acron_title_values,linestyle='--',color='r', marker='^',label='GCSE Acronym+Title')
plt.plot(x_values,gcse_acron_title_cordis_values,linestyle='--',color='g', marker='>',label=' GCSE Acronym+Title-Cordis')
plt.plot(x_values,gcse_acron_title_cordis_ec_values,linestyle='--',color='b', marker='v',label='GCSE Acronym+Title-Cordis-EC')
plt.grid()
plt.legend()
plt.xlabel('Number of Top Results')
plt.ylabel('URL Recover Performance')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
