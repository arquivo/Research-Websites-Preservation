# Mementos

```python
df = pd.read_csv('mementos_golden_dataset.csv')
```

```python
import re

regex = re.compile('\d{14}')

df.fillna('', inplace=True)
for index,row in df[['memento-closest']].iterrows():    
    if row['memento-closest'] != '':
        archive = row['memento-closest'].split('/')[2]        
        time = regex.search(row['memento-closest'])
        if time != None:
            time = regex.search(row['memento-closest']).group(0)
    else:
        time = ''
        archive = 'Not Archived'
        
    df.ix[index,'webarchive'] = archive
    df.ix[index,'timestamp'] = time    
```
